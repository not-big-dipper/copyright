```python
import aiohttp
import io

BASE_URL = 'http://176.109.105.24:8000/'

async def upload_video(filename: str, data: bytes, group):
    url = BASE_URL + 'video'
    params = {
        'title': filename,
        'description': filename,
        'group': group
    }

    async with aiohttp.ClientSession() as session:
        async with session.post(url, data={'file': data}, params=params) as resp:
            return await resp.json()
```

```python
import zipfile
import asyncio
import os
from tqdm import tqdm

INDEX_ZIPFILE = '/home/teexone/lct/14. RUTUBE/index.zip'

coros = []
filenames = []
with zipfile.ZipFile(INDEX_ZIPFILE, 'r') as zip_ref:
    for filename in tqdm(zip_ref.namelist()):
        if filename.endswith('.mp4'):
            with zip_ref.open(filename) as file:
                filename = os.path.basename(filename)
                data = file.read()
                coros.append(upload_video(filename, data, 'index'))
                filenames.append(filename)
        if len(coros) == 10:
            await asyncio.gather(*coros)
            uploaded.update(filenames)
            coros = []
            filenames = []

if coros:
    await asyncio.gather(*coros)
    uploaded.update(filenames)
    
print('Uploaded', len(uploaded), 'files')
```
